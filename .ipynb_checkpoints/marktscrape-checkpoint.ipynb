```python
def team_lookup(name):
    team_list = [
        'bournemouth',
        'arsenal', 
        'aston villa', 
        'birmingham city', 
        'blackburn rovers',
        'blackpool',
        'bolton wanderers', 
        'brighton hove albion', 
        'burnley', 
        'cardiff city', 
        'charlton athletic', 
        'chelsea', 
        'crystal palace', 
        'derby county', 
        'everton', 
        'fulham', 
        'huddersfield town',
        'hull city',
        'leicester city', 
        'liverpool', 
        'manchester city', 
        'manchester united', 
        'middlesbrough', 
        'newcastle united', 
        'norwich city', 
        'portsmouth', 
        'queens park rangers', 
        'reading', 
        'sheffield united', 
        'southampton', 
        'stoke city', 
        'sunderland',
        'swansea city',
        'tottenham hotspur', 
        'watford', 
        'west bromwich albion', 
        'west ham united', 
        'wigan athletic', 
        'wolverhampton wanderers'
    ]
    max_score = 0
    name_to_ret = ''
    for team in team_list:
        temp_score = fuzz.ratio(name, team)
        if temp_score > max_score:
            max_score = temp_score
            name_to_ret = team
        
    return name_to_ret
```

```python

def build_aux_table(year):
    season = build_by_year(year)
    for team in season:
        team.append(get_squad_info(team[1]))
    #print(season)
    name = []
    link = []
    g_age = []
    g_total_value = []
    g_avg_value = []
    d_age = []
    d_total_value = []
    d_avg_value = []
    m_age = []
    m_total_value = []
    m_avg_value = []
    a_age = []
    a_total_value = []
    a_avg_value = []
    for team in season:
        name.append(team[0])
        link.append(team[1])
        g_age.append(team[2][0][0])
        g_total_value.append(team[2][0][1])
        g_avg_value.append(team[2][0][2])
        d_age.append(team[2][1][0])
        d_total_value.append(team[2][1][1])
        d_avg_value.append(team[2][1][2])
        m_age.append(team[2][2][0])
        m_total_value.append(team[2][2][1])
        m_avg_value.append(team[2][2][2])
        a_age.append(team[2][3][0])
        a_total_value.append(team[2][3][1])
        a_avg_value.append(team[2][3][2])
    
    df = pd.DataFrame({'name': pd.Series(name), 'year': year,'link': pd.Series(link), 'g_age': g_age, 'g_total_value':g_total_value,
                  'g_avg_value': g_avg_value, 'd_age': d_age, 'd_total_value':d_total_value,
                  'd_avg_value': d_avg_value, 'm_age': m_age, 'm_total_value':m_total_value,
                  'm_avg_value': m_avg_value, 'a_age': a_age, 'a_total_value':a_total_value,
                  'a_avg_value': a_avg_value})
    
    df['name'] = df['name'].apply(lambda x: team_lookup(x))
    
    return df
    
```

```python
# returns 2 dicts, 1 of spent, 1 of sales
def get_all_transfer_info_year(url, year):
    date_dict = {
    '18/19': '2018',
    '17/18': '2017',
    '16/17': '2016',
    '15/16': '2015',
    '14/15': '2014',
    '13/14': '2013',
    '12/13': '2012',
    '11/12': '2011',
    '10/11': '2010',
    '09/10': '2009',
    '08/09': '2008',
    '07/08': '2007',
    '06/07': '2006',
    '05/06': '2005',
    '04/05': '2004'}
    year_to_ret = ''
    for dash, whole in date_dict.items():
        if year == whole:
            year_to_ret = dash
    fixed_url = url.replace("startseite", "alletransfers")[1:]
    full_url = 'https://www.transfermarkt.co.uk/' + fixed_url 
    tree = requests.get(full_url, headers = headers)
    s = BeautifulSoup(tree.content, 'html.parser')
    incoming = s.select("td.redtext")
    in_and_out = []
    for idx, spent in enumerate(incoming):
        for parent in incoming[idx].parents:
            if parent.get('class') == ['box']:
                split = parent.text.split()
                if split[0] == 'Arrivals' and split[1] == year_to_ret:
                    in_and_out = [spent.text, '']
                    break
            
        # if we got here, we didnt find it!
    if not bool(in_and_out):
        in_and_out = ['0', '']
    outgoing = s.select("td.greentext")    
    for idx, sale in enumerate(outgoing):
        for parent in outgoing[idx].parents:
            if parent.get('class') == ['box']:
                split = parent.text.split()
                if split[0] == 'Departures' and split[1] == year_to_ret:
                    in_and_out[1] = sale.text
    
    if in_and_out[1] == '':
        in_and_out[1] =  '0'
    return in_and_out
                
#get_all_transfer_info_year('/fc-liverpool/startseite/verein/31/saison_id/2018', 2018)



def get_all_transfer_info(url):
    date_dict = {
    '18/19': '2018',
    '17/18': '2017',
    '16/17': '2016',
    '15/16': '2015',
    '14/15': '2014',
    '13/14': '2013',
    '12/13': '2012',
    '11/12': '2011',
    '10/11': '2010',
    '09/10': '2009',
    '08/09': '2008',
    '07/08': '2007',
    '06/07': '2006',
    '05/06': '2005',
    '04/05': '2004'}
    
    
    fixed_url = url.replace("startseite", "alletransfers")[1:]
    full_url = 'https://www.transfermarkt.co.uk/' + fixed_url 
    tree = requests.get(full_url, headers = headers)
    s = BeautifulSoup(tree.content, 'html.parser')
    incoming = s.select("td.redtext")
    in_and_out = {}
    for idx, spent in enumerate(incoming):
        for parent in incoming[idx].parents:
            if parent.get('class') == ['box']:
                split = parent.text.split()
                print(split[0] + ' split')
            
                if split[1] == '03/04':
                    break
                if split[1] != '19/20':
                    if split[1] not in in_and_out:
                        if spent.text != '':
                            in_and_out[date_dict[split[1]]] = [spent.text, '0']
                        else:
                            in_and_out[date_dict[split[1]]] = ['0m']
        if split[1] == '03/04':
            break
    outgoing = s.select("td.greentext")
    yearly_sale = []
    for idx, sale in enumerate(outgoing):
        for parent in outgoing[idx].parents:
            if parent.get('class') == ['box']:
                split = parent.text.split()
                
                print(split[0] + ' split')
                if split[1] in date_dict:
                    if date_dict[split[1]] in in_and_out:
                        if spent.text != '':
                            in_and_out[date_dict[split[1]]][1] = sale.text
                        else:
                            in_and_out[date_dict[split[1]]].append('0m')
    
    
    print(in_and_out.items())
    return [ full_url, dict(list(in_and_out.items())[:15])]
       
    
    
    
def get_all_transfer_info2(url):
    date_dict = {
    '18/19': '2018',
    '17/18': '2017',
    '16/17': '2016',
    '15/16': '2015',
    '14/15': '2014',
    '13/14': '2013',
    '12/13': '2012',
    '11/12': '2011',
    '10/11': '2010',
    '09/10': '2009',
    '08/09': '2008',
    '07/08': '2007',
    '06/07': '2006',
    '05/06': '2005',
    '04/05': '2004'}


    fixed_url = url.replace("startseite", "alletransfers")[1:]
    full_url = 'https://www.transfermarkt.co.uk/' + fixed_url 
    data = requests.get(full_url, headers = headers)
    soup = BeautifulSoup(data.content, "lxml")
    #incoming = s.select("td.redtext")
    divs = soup.find_all('div', attrs={'class':'table-header'})
    for div in divs:
        print(div)
        print('----- text -----')
        print(div.text.rstrip())
    print(div)
    in_and_out = {}
    for key, value in date_dict.items():
        print(key, value)
        arrive = 'Arrivals ' + key
        #print(arrive)
        
        #arrival = s.select("div.box").select("div.table-header")
        #arrival = s.find("div.b")#, string='Arrivals 17/18')
        #print(arrival)
    
    #print(in_and_out.items())
    #return [ full_url, dict(list(in_and_out.items())[:15])]
```

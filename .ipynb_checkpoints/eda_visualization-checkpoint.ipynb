```python
team_arr = []
for team in TEAM_LIST:
    file = 'teams/' + team + '.csv'
    df = pd.read_csv(file, index_col=0)
    df['name'] = team
    team_arr.append(df)

all_year = pd.concat(team_arr)
all_year.sort_values(['year', 'Rank'], ascending=[False, True], inplace=True)
```

```python
all_year['squad_age'] = all_year[['g_age', 'd_age', 'm_age', 'a_age']].mean(axis=1)
all_year['squad_total_value'] = all_year[['g_total_value', 'd_total_value', 'm_total_value', 'a_total_value']].mean(axis=1)
all_year['squad_avg_value'] = all_year[['g_avg_value', 'd_avg_value', 'm_avg_value', 'a_avg_value']].mean(axis=1)
all_year['net_spent_transfer'] = all_year['sold'] - all_year['bought'] 

```

```python
import scipy.stats as stats
import statsmodels.stats.weightstats as ws
```

```python
import random
def make_random_seasons(df, func, size=1):
    to_ret = []
    for i in range(size):
        season_list = []
        for rank in range(1, 21):
            season = random.choice(SEASON_LIST)
            df_year = df[df['year'] == season]
            row = df_year[df_year['Rank'] == rank]
            season_list.append(row)
        df_season = pd.concat(season_list)
        df_season['gen_year'] = i
        df_season.sort_values(['Points', 'GD'], ascending=False, inplace=True)
    
        values = list(range(1, 21))
        data = np.array(values)
        s = pd.Series(data)
        
        df_season['reRank'] = s.values
        to_ret.append(df_season)
        
    return pd.concat(to_ret)
        

    
```

```python
gen_seasons = make_random_seasons(all_year, 1000)
gen_seasons.head()
```
